```python
cat_cols = ["type_of_meal_plan","required_car_parking_space","room_type_reserved","arrival_year","arrival_month","arrival_date","market_segment_type","repeated_guest"]
num_cols = ["no_of_adults","no_of_children","no_of_weekend_nights","no_of_week_nights","lead_time","no_of_previous_cancellations","no_of_previous_bookings_not_canceled","avg_price_per_room","no_of_special_requests"]
```

```python
# detecting ouliers for all before removing 
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3-Q1

before_rvm_detect_outliers = ((df[num_cols]<(Q1- 1.5*IQR)) | (df[num_cols]>(Q3+ 1.5*IQR))).sum()
before_rvm_detect_outliers
```

```python
detected_outlier_cols = ["no_of_adults","no_of_children","no_of_weekend_nights","no_of_week_nights","lead_time","no_of_previous_cancellations","no_of_previous_bookings_not_canceled","avg_price_per_room","no_of_special_requests"]

Q1 = df[detected_outlier_cols].quantile(0.25)
Q3 = df[detected_outlier_cols].quantile(0.75)
IQR = Q3-Q1

#detect_outliers = ((titanic[detected_outlier_cols]<(Q1- 1.5*IQR)) | (titanic[detected_outlier_cols]>(Q3+ 1.5*IQR))).sum()
df = df[~((df[detected_outlier_cols] < (Q1 - 1.5 * IQR)) |(df[detected_outlier_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
# detecting ouliers for all before removing 
after_rvm_detect_outliers = ((df[num_cols]<(Q1- 1.5*IQR)) | (df[num_cols]>(Q3+ 1.5*IQR))).sum()
after_rvm_detect_outliers
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# fitting the X_train[num_cols] in standardization model
scaler = StandardScaler()
scaler = scaler.fit(X_train[num_cols])
```

```python
# fitting the X_train[num_cols] and X_test[num_cols] in standardization model
X_train_std = pd.DataFrame(scaler.transform(X_train[num_cols]), columns = X_train[num_cols].columns)
X_val_std = pd.DataFrame(scaler.transform(X_val[num_cols]), columns = X_val[num_cols].columns)
```

```python
# creationg dummies for cat_cols from X_train
X_train_dummies = pd.get_dummies(X_train[cat_cols],drop_first = True)
```

```python
X_val_dummies = pd.get_dummies(X_val[cat_cols],drop_first = True)
```

```python
X_train_dummies = X_train_dummies.reset_index()
X_train_dummies
```

```python
X_val_dummies = X_val_dummies.reset_index()
X_val_dummies
```

```python
final_X_train = pd.concat([X_train_std,X_train_dummies],axis = 1)
final_X_train
```

```python
from sklearn.svm import SVC
```

```python
# model building by SVM algorithm
svm = SVC(kernel="linear", C = 10)
svm = svm.fit(final_X_train,y_train)
```

```python
svm
```

```python
# predicting the y value for the train data
svm_y_train_preds = svm.predict(final_X_train)
svm_y_train_preds[0:10]
```

```python
# predicting the y value for the val data
svm_y_val_preds = svm.predict(final_X_val)
svm_y_val_preds[0:10]
```
